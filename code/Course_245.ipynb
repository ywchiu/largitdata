```python
import nest_asyncio
import asyncio
from pyppeteer import launch

nest_asyncio.apply()

async def scrape_website(url):
    # 啟動瀏覽器
    browser = await launch(headless=True)
    page = await browser.newPage()

    # 開啟指定的網頁
    await page.goto(url)

    # 等待網頁載入完成
    await page.waitForSelector('body')

    # 提取網頁的內容
    content = await page.content()

    # 關閉瀏覽器
    await browser.close()

    return content

# 主程序
url = 'https://openai.com/index/introducing-structured-outputs-in-the-api/'
content = asyncio.get_event_loop().run_until_complete(scrape_website(url))

# 輸出內容
print(content)
```

```python
import nest_asyncio
import asyncio
from pyppeteer import launch
from pyppeteer_stealth import stealth 

nest_asyncio.apply()

async def scrape_website(url):
    # 啟動瀏覽器
    browser = await launch(headless=True)
    page = await browser.newPage()

    await stealth(page)
    
    # 開啟指定的網頁
    await page.goto(url)

    # 等待網頁載入完成
    await page.waitForSelector('body')

    # 提取網頁的內容
    content = await page.content()

    # 關閉瀏覽器
    await browser.close()

    return content

# 主程序
url = 'https://openai.com/index/introducing-structured-outputs-in-the-api/'
content = asyncio.get_event_loop().run_until_complete(scrape_website(url))

# 輸出內容
print(content)
```
